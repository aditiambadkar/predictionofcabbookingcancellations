Data Pre-processing

```python
#checking for null values
data.isnull().sum()
```

```python
#data cleaning: removing unnecessary columns
data = data.drop(['package_id', 'to_area_id', 'from_area_id', 'from_city_id','to_city_id', 'to_date', 'to_lat', 'to_long', 'from_lat', 'from_long', 'id', 'user_id'], axis = 1)
```

```python
import numpy as np
a = np.array(data['from_date'])
a = a.astype('str')

#separate date and time for attribute 'from date'
t_date = [] #trip date

for j in a:
    t_date.append(j.split()[0])
```

```python
t_time = [] #trip time
for i in a:
    t_time.append(i.split()[1])
```

```python
t_day = []
t_month = []
t_year = []

for q in t_date:
    t_day.append(q.split('/')[1])
    t_month.append(q.split('/')[0])
    t_year.append(q.split('/')[2])
```

```python
t_day
```

```python
t_month
```

```python
t_hours = []
t_min = []

for q in t_time:
    t_hours.append(q.split(':')[0])
    t_min.append(q.split(':')[1])
```

```python
a = np.array(data['booking_created'])
a = a.astype('str')

b_date = [] #booking date

#separate date and time for attribute 'booking created'
for j in a:
    b_date.append(j.split()[0])

b_time=[]
for i in a:
    b_time.append(i.split()[1])
```

```python
b_day = []
b_month = []
b_year = []

for q in b_date:
    b_day.append(q.split('/')[1])
    b_month.append(q.split('/')[0])
    b_year.append(q.split('/')[2])

b_hours = []
b_min = []

for q in b_time:
    b_hours.append(q.split(':')[0])
    b_min.append(q.split(':')[1])
```

```python
#computing difference in days between 'booking created' and 'from date'
import datetime

difference_days = []
trip_date = []
booking_date = []

for i in range(len(b_hours)):
    trip_date.append(datetime.datetime(int(t_year[i]), int(t_month[i]), int(t_day[i]), int(t_hours[i]), int(t_min[i])))
    booking_date.append(datetime.datetime(int(b_year[i]), int(b_month[i]), int(b_day[i]), int(b_hours[i]), int(b_min[i])))
    
    diff = trip_date[i] - booking_date[i]
    days = diff.days
    
    difference_days.append(days)
```

```python
data.info()
```

```python
data['days_difference'] = np.array(difference_days)
```

```python
data = data[data.days_difference != -1]
```

---

Independent variables X (input) and dependent variable y (target)

---

```python
X = data.drop(['Car_Cancellation'], axis = 1)
y = data['Car_Cancellation']
```

---

Split the dataset into training and test sets

---

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3) # 70% training and 30% test data
```

---

Random Forest Classifier

---

```python
from sklearn.ensemble import RandomForestClassifier

#create a Random Forest Classifier
clf = RandomForestClassifier(n_estimators = 100)

#train the model using the training sets
clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)
```

```python
from sklearn import metrics

#model accuracy, how often is the classifier correct?
print("Accuracy:", metrics.accuracy_score(y_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix
cnf_matrix = confusion_matrix(y_test, y_pred)
print(cnf_matrix)
print(clf.score(X_test, y_test))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

Gaussian Naive Bayes Classifier

```python
from sklearn.neighbors import KNeighborsClassifier

#create a KNN classifier
knn = KNeighborsClassifier(n_neighbors = 50)

#train the model using the training sets
knn.fit(X_train, y_train)

#predict the response for test dataset
y_pred2 = knn.predict(X_test)
```
